```python
import matplotlib.pyplot as plt
import numpy as np
import os

values = {(1,1):0.151, (1,5):0.179, (1,10):0.196, (1,20):0.215, (5,1):0.165, (5,10):0.215,
          (5,20):0.213, (10,1):0.181, (10,5):0.199, (10,20):0.207, (20,1):0.202, (20,5):0.194, (20,10): 0.205}


a = np.zeros([20,20])
for k,v in values.items():
    a[k[0]-1, k[1]-1]=v

plt.imshow(a, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()
```
